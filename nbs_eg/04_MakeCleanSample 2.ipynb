## Calculating LINEAR periods

```python
# recompute periods using nterms=3
LINEAR_data = calculating_period('LINEAR', 3, 350, False)
```

```python
# we need periods and IDs
PeriodsLinear = np.array(LINEAR_data['Period'])
IDs = [x for x in data.ids]
```

```python
# now get original periods, as were used in e.g. fig. 10.20, see
# https://www.astroml.org/book_figures/chapter10/fig_LINEAR_clustering.html
# NOTE: there are only 6146 stars in this sample, not 7010! 
#       AND: for 942 stars, IDs do not match so the overlap is 5204 stars
from astroML.datasets import fetch_LINEAR_geneva
dataPeriods = fetch_LINEAR_geneva() 
PlinearOrig = 10**(dataPeriods['logP'].T)
raOrig = dataPeriods['ra'].T
decOrig = dataPeriods['dec'].T
idOrig = list(dataPeriods['LINEARobjectID'].T)
```

```python
ibad = igood = 0
goodPeriodOrig = []
goodPeriodNew = []
goodIDs = []
goodOrigIndex = []
# loop over 7010 stars from astroML distribution
for id in idOrig: 
    if id in IDs:
        # this star exists in the "geneva" subsample
        index = IDs.index(id)
        goodPeriodOrig.append(PlinearOrig[idOrig.index(id)])
        goodPeriodNew.append(PeriodsLinear[index])
        goodIDs.append(id)
        goodOrigIndex.append(idOrig.index(id))
        igood += 1
    else: 
        # not in the "geneva" list 
        ibad += 1
PeriodOrig = np.array(goodPeriodOrig)
PeriodNew = np.array(goodPeriodNew)
print(igood, ibad)  
```

```python
# now save these 5204 for later perusal in a master file (plus metadata, such as colors)
fout = open('LINEAR_Periods_nterm3.txt', "w")
fout.write("    LINEAR ID     Porig         Pnew")
fout.write("        ra             dec          ug      gi      iK      JK       logP       Ampl    skew    kurt  magMed nObs LCtype \n")    
for i in range(0,len(goodIDs)):
    ID = goodIDs[i]
    Porig = goodPeriodOrig[i]
    Pnew = goodPeriodNew[i]
    s = str("%12s " % ID) + str("%12.8f  " % Porig) + str("%12.8f  " % Pnew)
    OrigIndex = goodOrigIndex[i]
    LID = dataPeriods[OrigIndex ]['LINEARobjectID']
    assert{ID == LID}
    ra = dataPeriods[OrigIndex]['ra']
    dec = dataPeriods[OrigIndex]['dec']
    s = s + str("%12.7f " % ra) + str("%12.7f  " % dec)
    for q in ['ug','gi','iK','JK']:
        s = s + str("%7.2f " % dataPeriods[OrigIndex][q])
    s = s + str("%12.8f " % dataPeriods[OrigIndex]['logP'])     
    for q in ['Ampl','skew','kurt','magMed']:
        s = s + str("%7.2f " % dataPeriods[OrigIndex][q])
    s = s + str("%4.0f " % dataPeriods[OrigIndex]['nObs'])     
    s = s + str("%2.0f " % dataPeriods[OrigIndex]['LCtype'])     
    s = s + "\n"
    fout.write(s)             
fout.close() 
```
